<h1>How did International K-Pop Fans React to Multinational K-Pop Group?</h1>

<font size=3>

What concept can sound more contradictory than a K-Pop group not based in Korea, having no Korean members, and mostly singing in English? Multinational K-Pop has been a controversial concept in the culture for a long time, but now their numbers are increasing. In this project, I focused on six Multinational K-Pop groups and discussed which aspects of Multinational did K-Pop fans concern about through an analysis of comments collected from these groups' music videos on YouTube.

 </font>

```python
from IPython.display import Image
from IPython.core.display import HTML 
```

<h2> Research Objective </h2>

<font size=3>
    
1. Which aspects of a Multinational K-Pop group did K-Pop fans concern about? What is their reaction?
2. Whether it is a good idea for an investor to promote a Multinational K-Pop now?
3. If someone wants to promote a Multinational K-Pop group, what should be paid the most attention to?

</font>

<h2> Research Subjects </h2>

<font size=3>
    
1. Exp Edition, consisting of four white Americans
2. UHSN, consisting of ten members from ten different countries
3. 5High, consisting of five Indians
4. KAACHI, consisting of three girls from the UK and one girl from Korea
5. Black Swan, consisting of three Korean, one Senegalese, and one Brazilian
6. Prisma, consisting of two Korean, one American, one Italian, and one Spanish

</font>

<h2> Data Collection </h2>

<font size=3>
    
1. Data collection was done through Google API. Parameter order was set to relevance, which means the API will fetch the most relevant comments based on the Google algorithm. Both comments and replies were fetched.
2. Data collection was done on September 6, 2021.
3. A total of sixteen thousand comments were collected from eight YouTube videos at the end of the data collection.

</font>

<h2> Key Findings </h2>

<font size=3>
    
1. Promoting Multinational K-Pop is still a good strategy to draw attention as the discussion about Multinational K-Pop did not seem to cease for the last few months even without any Multinational K-Pop having promotions.
2. This analysis provides a robust prediction model for Multinational K-Pop group managers to monitor their future promotions.
3. The result helps identify the aspect of existing Multinational K-Pop groups that need to be worked on more to satisfy their viewers.
    
</font>

<h2> Result Three: Which Aspects of a Multinational K-Pop group did K-Pop Fans Concern About? </h2>

<font size=3>
    
The following analysis, focusing only on comments written in English because of the researcher's literacy, was based on the result of a classification model using TF-IDF and Naive Bayes. Three mutually inclusive variables were made after initial data exploration. These variables are comments about the quality of the K-Pop group (quality), nationalist or ethnicist comments (nationalist-ethnicist), and comments that provide concepts about what K-Pop is (k-pop).
    
</font>

```python
Image(url= "https://chunjuihsu.github.io/kpop_project/story_telling/classification_count.png", width=800, height=600)
```

<font size=3>
    
1. The figure represents the count of each classification predicted by the TF-IDF and Naive Bayes model.
2. The small result sizes can be a consequence of the small train-test dataset (400 in an 8000 dataset).
    
</font>

![languageuse](4.sample_comments.png)

<font size=3>
    
1. The above snippet presents three sample comments from each classification.
2. It can be seen the prediction worked fairly well, but the drawback of TF-IDF (essentially bag of words) was also shown as the model misclassified sentences such as "you can do better" as comments about the quality of a K-Pop group, and the fact that the model cannot tell the difference between Korean of the language and Korean of the ethnicity.
    
</font>

![tfidf_quality](5.tfidf_model_quality.png)

<font size=3>
    
1. This figure presents the mean TF-IDF score of each word scored in the process for the model to predict whether a sentence is a comment about quality or not.
2. It can be seen the model works quite well as words that refer to different aspects of a K-Pop group, such as song, vocal, and music, were as expected on the left-hand figure and not on the right-hand one.
    
</font>

![tfidf_nationalist_ethnicist](6.tfidf_model_nationalist_ethnicist.png)

<font size=3>
    
1. This figure presents the mean TF-IDF score of each word scored in the process for the model to predict whether a sentence is a nationalist and ethnicist comment or not.
2. It can be seen the model works quite well as words such as race, asian, white, and black, were predicted to be words of nationalist and ethnicist comments.
    
</font>

![tfidf_quality](7.tfidf_model_kpop.png)

<font size=3>
    
1. This figure presents the mean TF-IDF score of each word scored in the process for the model to predict whether a sentence is a comment that provides an idea about what K-Pop is or not.
2. From the words placed on the left-hand figure but not on the right-hand figure, we found some unique words that are predicted as words describing K-Pop. These words are company and training.
3. The model also included many other words that are also words about quality, such as song and music, and nationalist or ethnicist words, such as korea and asian. This observation suggested the discussion about K-Pop was rarely made without mentioning its quality and its nationalist-ethnicist background.
4. To verify this observation, I did the following network analysis.
    
</font>

![network](8.network_of_concepts.png)

<font size=3>

1. This figure represents the connections between concepts that were identified through the help of the prediction models. Each node represents a concept, which usually consists of a lexicon that shares this concept (e.g. voice and vocal as the lexicon of voice). The width of edges that link different nodes represents the count of the co-occurrence of the two concepts in the same sentences.
2. The figure shows that K-Pop had a strong association with concepts about quality in volume. K-Pop particularly co-occurred with the idea of song many times, implying songs are an important part of K-Pop to most K-Pop fans.
3. For nationalist and ethnicist concepts, the idea of Asians was associated with K-Pop more often than others.
4. While this figure only provides frequencies of two given concepts that were co-occurred, this figure did not present whether it was a positive association or a negative association. For example, "K-Pop does not need to be Asian." and "K-Pop is for Asian!!!" could mean very different things, but both occurrences will be calculated in this network.
5. Therefore, this chapter concludes that the quality is more important to K-Pop than its nationalist or ethnicist background, but how fans actually felt about the connection will require further sentiment analysis to tell.

</font>

<h2> Result Four: An Aspect-Based Sentiment Analysis of Multinational K-Pop</h2>

<font size=3>
    
The following sentiment analysis used a pre-trained lexicon- and rule-based sentiment analyzer Vader to determine the sentiment score at the scale of a sentence. It focused on aspects rather than a general sentiment score, so it is more insightful for group managers to identify the problem and design business strategies accordingly.
    
</font>

![qualitysentiment](9.sentiment_distribution_quality.png)

<font size=3>
    
1. The figure shows the proportions of different sentiments reflected in the sentences about quality under music videos of different Multinational K-Pop groups.
2. The result shows that 5 High had the greatest proportion of negative sentiment, around 21.5%, followed by KAACHI's 19.6%, suggesting there was a huge concern about their quality.
3. Meanwhile, Prisma and Black Swan had a smaller proportion of negative sentiment about their quality, suggesting fans were satisfied with their music and performance.
    
</font>

![negativequality](10.negative_distribution_quality.png)

<font size=3>
    
1. This figure shows the proportion of negative sentiment about different aspects of quality. The X-axis represents the percentage of a given negative sentiment toward a specific quality aspect occupied in the total negative sentiment about quality. The different colors of the circle represent the different aspects the circles represent, and the size of the circle indicates the count of such occurrences. 
2. The result shows that songs are the main concern most commenters had about these multinational K-Pop groups, suggesting a place where the group managers can work on more in their next release.
3. Vocal ability appeared to be a big problem for 5 High, as more than 35% of the negative sentiment about quality, addressing concerns about the vocal aspect.
4. Black Swan and Prisma performed relatively well regarding their language, dance, and vocal ability. It can be seen from the small circular shapes that represent the count of negative sentiments toward these qualities.
    
</font>

![nesentiment](11.sentiment_distribution_nationalist_ethnicist.png)

<font size=3>
    
1. The figure presents the proportion of different sentiments reflected in the sentences having a nationalist and ethnicist utterance under music videos of Multinational K-Pop groups.
2. The result shows that UHSN had the greatest proportion of negative sentiment, around 34.0%, followed by KAACHI's 30.1%, suggesting there was a big concern about their nationalist and ethnicist background.
3. Meanwhile, more than 45% of the sentiment with a nationalist and ethnicist utterance under the music videos of 5 High and Prisma were positive, suggesting their nationalist-ethnicist backgrounds appealed to many.
    
</font>

![negativene](12.negative_distribution_nationalist_ethnicist.png)

<font size=3>
    
1. This figure presents the proportion of different sentiments about different aspects of nationalist-ethnicist utterance. The X-axis represents the percentage of a given sentiment toward occupied in all utterances. The different colors of the circle represent the different sentiments, and the size of the circular shape indicates the count of such occurrences. 
2. The result shows that Indian was the nationalist-ethnicist concept that was the most likely to be positively addressed. Around 46 percent of the utterances were addressed positively.
3. Meanwhile, White was the concept that was the most likely to be negatively addressed. More than 35% of the sentences that mentioned White were negatively addressed.
3. In general, racial issues are topics likely to be addressed negatively by K-Pop fans, but Black, Indian, and Asian appeared to be safer situations than White.
    
</font>

<h2> Conclusion </h2>

<font size=3>

1. Considering the unceasing discussion about Multinational K-Pop, this analysis suggested promoting Multinational K-Pop is still an effective strategy to draw attention for newly promoted K-Pop groups.
2. This analysis found that there were more comments about the performative quality of Multinational K-Pop than comments about the nationalist and ethnicist aspect of Multinational K-Pop. It suggested what most K-Pop fans were concerned about Multinational K-Pop was their performative quality rather than their nationalist and ethnicist backgrounds.
3. The analysis result identified the aspects of quality that Multinational K-Pop groups need to work on more to satisfy their viewers. Songs are the shared aspect that most of them will need to work on more. 5 High needs to improve their vocal ability, and KAACHI might need to start with their dancing and vocal ability.
4. The prediction models were evaluated to be robust, so they can be used to facilitate the future promotion of Multinational K-Pop groups.
5. Ideally, a Multinational K-Pop manager would like to have more viewers focusing on the group's performative qualities rather than their nationalist-ethnicist background because nationalist-ethnicist comments were more likely to generate negative sentiments.
6. Nevertheless, Black and Indian K-Pop idols appeared to be a safer choice than White idols because of the greater proportion of positive sentiments generated because of them.
7. The analysis also found that the members' nationality helps generate traffics from the national market of that given member.
8. Therefore, this project proposed that a non-White idol, who is from a country with a large population and consumption power, would be an ideal option for the next Multinational K-Pop to promote.
